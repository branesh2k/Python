## Basic Concepts

Matplotlib is built around two primary interfaces:

1. `Pyplot API:` A state-based interface, similar to MATLAB, that provides a convenient way to create and customize plots.  This is the most commonly used interface, especially for quick and simple visualizations.

2. `Object-Oriented API:` Offers more control and customization by directly manipulating figure and axis objects.  This approach is more powerful and flexible, suitable for complex plots and applications.

### Basic Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Sample data
x = [1,2,3,4,5,6]
y = [2,3,5,7,9,11]

#create a plot
plt.plot(x,y)

#labelling axes and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("sample graph")

#show the plot
plt.show()
```

### customization
Changing Colors, Line Styles, and Markers

#### Colors
*  Matplotlib supports a variety of ways to set colors, including named colors, hexadecimal color codes, RGB tuples, and predefined color maps.

```python
plt.plot(x, y, color='green')          # Using named color
```

```python
plt.plot(x, y, color='#FF5733')        # Using hex color code
```

```python
plt.plot(x, y, color='tab:blue')       # Using color map
```

```python
plt.plot(x, y, color=(1.0, 0.2, 0.3))  # Using RGB tuple
```

#### Line-style
* Customize the appearance of lines using line styles like solid, dashed, dash-dot, and dotted.

```python
plt.plot(x,y,linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

#### Markers
* Change the appearance of data points using different marker styles.

```python
plt.plot(x,y, marker='o')
```

```python
plt.plot(x,y,marker='s')
```

#### Modifying axes and lables

`Axis Limits:` Define the range of values displayed on the axes.

```python
plt.xlim(1,10)
plt.ylim(2,12)
plt.plot(x,y)
```

`Ticks and Tick Labels:` Customize the placement and appearance of ticks and their labels.

```python
plt.xticks([0, 2, 4, 6, 8, 10], ['0', 'Two', 'Four', 'Six', 'Eight', 'Ten'])
plt.yticks([0, 5, 10, 15, 20])

plt.grid(True)      #  for grid lines

plt.plot(x, y)
```
