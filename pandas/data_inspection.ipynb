### Data Cleaning

````python
df.info()
````

````python
type(df.job_posted_date[0])
````

* Here column `job_posted_date` has `object` dtype which is a string
* coverting to date_time.

````python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
````

````python
df.info()
````

````python
df['job_posted_month'] = df.job_posted_date.dt.month
df.info()
````

````python
df.sort_values('job_posted_date',inplace=True)
df
````

````python
df.drop(labels='salary_hour_avg',axis=1,inplace=True)
````

````python
df
````

````python
df.dropna(subset=['salary_year_avg'],inplace=True)
````
