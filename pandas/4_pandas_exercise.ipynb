### Question

Perform an exploratory analysis focusing on data jobs in india. Look at the average yearly salary, job count, and salary range (minimum and maximum salaries).

```python
import pandas as pd
from datasets import load_dataset

# Loading dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# data cleanup
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

Filtering based on country `India`

```python
india_jobs = df[df['job_location'] == 'India']
india_jobs
```

Before calculating average ,I'll remove `NaN`

```python
india_jobs=india_jobs[india_jobs['salary_year_avg'].notna()]
# (or) india_jobs.dropna(subset='salary_year_avg')
```

Calculating job count, avg salary and salary range in India and sorting it by average.

```python
india_jobs.groupby('job_title_short')['salary_year_avg'].agg(['count', 'min','max','median']).sort_values('median',ascending=False)
```
